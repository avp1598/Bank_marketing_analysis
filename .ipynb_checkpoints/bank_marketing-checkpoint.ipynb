<h1>Importing libraries</h1>

```python
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
np.random.seed(0) #setting random seed 
```

<h1>Prediction</h1>

```python
#dividing data into target and features
y=data['y']
X=data.drop(['y'],axis=1)
```

<p style="font-size:16px"> We are going to use the random forest classifier(<a href="https://towardsdatascience.com/the-random-forest-algorithm-d457d499ffcd">Random Forest)</a> to model our dataset. This model requires all the attributes to be in integer form, hence we have to convert the categorical data to numeric data. We do this by using the following pandas functions. </p>

```python
y=pd.get_dummies(y)
X=pd.get_dummies(X)
```

<p style="font-size:16px"> Next we split our dataset into training and testing sets. We use sklearn's built in funtion for this purpose </p>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

<p style="font-size:16px">Lets use the sklearn's random forest module to fit our data and find out its cross validation accuracy</p>

```python
clf=RandomForestClassifier(n_estimators=3,max_depth=11)
clf.fit(X_train, y_train)
scores = cross_val_score(clf, X_train, y_train, cv=5)
print("Mean Accuracy:",scores.mean())
```

<p style="font-size:16px">We used random parameters 3 and 11 for the classifier, Lets try to find out better parameters using sklearn's GridSearchCV funtion.<br><br>
Running the below cell will take a lot of time as it is iterating over all values in the range to find the best hyperparameters. I have ran it in my machine and the best parameters it returned was<br>
{'criterion': 'gini', 'max_depth': 7, 'max_features': 'auto', 'n_estimators': 13}<br><br>
So you can skip to the next cell.</p>

```python
param_grid = { 
    'n_estimators': list(range(2,20)),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : list(range(2,20)),
    'criterion' :['gini', 'entropy']
}
clf=RandomForestClassifier()
CV_rfc = GridSearchCV(estimator=clf, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
print(CV_rfc.best_params_)
```

<p style="font-size:16px">Lets run the classifier again with new hyperparameters</p>

```python
clf=RandomForestClassifier(n_estimators=13,max_depth=7)
clf.fit(X_train, y_train)
scores = cross_val_score(clf, X_train, y_train, cv=5)
print("Mean Accuracy:",scores.mean())
```

<p style="font-size:16px">That's amazing our model accuracy has crossed 90%.<br>
This completes our prediction phase</p>

<h1>Application </h1>

<p style="font-size:16px">We need to reduce the company's marketing cost by X% and acquire Y% of the prospects (compared to random calling), where X and Y are to be maximized</p>

<p style="font-size:16px"> Lets assume the cost of making calls is 1\$ and the profit gained from the client subscription is 20$<br>
    Hence profit Y gained from random calling : (4640)*20-41188</p>

```python
profit=len(data.loc[data['y']=='yes'])*20-len(data)
print(profit,"$")
#profit per call
ppc=profit/len(data)
ppc
```

<p style="font-size:16px">Now our model has an accuracy of 90% of predicting whether the client will buy the subscription or not.<br>
    Lets the say we have a new dataset of 50000 clients, suppose our model predicts 6000 of those clients will buy the subscription. Since our model has 90% accuracy 0.9*6000=5400 clients will actually buy and rest will ignore. </p>

```python
new_set=50000
predicted=6000
actual=5400
new_profit=predicted*20-actual
print(new_profit,"$")
#profit per call
n_ppc=new_profit/predicted
n_ppc
```

<p style="font-size:16px"> As you can see our profit per call has increased to 19.1 </p>

```python
#percentage increase
print(((n_ppc-ppc)/(ppc))*100,'%')
```
